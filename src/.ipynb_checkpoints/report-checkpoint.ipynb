El analisis de la tendencia de la movilidad se realizara sobre los datos extraidos directamente del sitio oficial de apple. 

```python
dfc= pd.read_csv("../data/apple_clean.csv")
```

```python
dfc.head()
```

>En la mayoria de las ocasiones el uso de los terminales para acceder al sistema de navegacion se realiza en un vehiculo. Por ello se centrara el estudio de la movilidad en ese entorno. 

```python
in_madrid = dfc['region'] == 'Madrid'
filter_madrid = dfc[in_madrid]
filter_madrid
```

La estructura de la base de datos exige ciertas manipulaciones previas 

- Creacion de filtros para precisar los distintos parametros de nuestra busquedad.

>filter_mad para localizar la ciudad.

>filter_es por si existiera otra ciudad llamada madrid.

>filter_geo_cit por si saliese otro registro ya que el metodo .contains() podria devolvernos cualquier palabara con esos caracteres 

>filter_drive para seleccionar el tipo de transporte 
 

```python
filter_mad = (dfc['region'].str.contains('Madrid'))
filter_es = (dfc['country'].str.contains('Spain'))
filter_geo_city = (dfc['geo_type'].str.contains('city'))
filter_drive = (dfc['transportation_type'].str.contains('driving'))
```

```python
#consulta conductores madrid ciudad 
consultaM = dfc[filter_mad & filter_drive & filter_geo_city & filter_es]
consultaM
```

- Generamos la serie haciendo la traspuesta ajustando la aparicion de la serie a partir del primer registro con fecha. 

```python
serie_m = np.transpose(consultaM)
serie_mdc = serie_m[6:]
serie_mdc
```

## Visualizacion de la movilidad de  la ciudad de madrid

Numero de habitantes de la ciudad de madrid

```python
es_df[es_df['Nombre'].str.contains('Madrid')]
```

### Pre-pandemia 

Los valores de la tendencia empiezan el dia lunes 13 de enero del 2020 en el valor 100, siendo este un punto de referencia para la correcta lectura de los valores. 

```python
pre_confinamiento = serie_mdc[:19]
```

```python
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
ax.plot(pre_confinamiento, 'blue', label= "movilidad")
plt.title('movilidad ')
ax.legend(loc=2)
plt.grid(True)
plt.show()
```

>Observamos cierto patron de crecimiento y decrecimiento, alcanzando el maximo los viernes y el minimo los domingos.  Por norma general nos desplazamos mas entre semana que los fines de semana. O por lo menso hacemos mas uso de nuestros dispositivos para llegar a nuestros destinos.

## Comparativa con Barcelona

```python
filter_bcn = (dfc['region'].str.contains('Barcelona'))
```

```python
#consulta conductores madrid ciudad 
consultaB = dfc[filter_bcn & filter_drive & filter_geo_city & filter_es]
consultaB
```

```python
# Plot of the Time Series
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
ax.plot(serie_bdc, 'green', label= "movilidad")
plt.title('movilidad ')
ax.legend(loc=2)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
ax.plot(serie_bdc, 'blue', label= "madrid")
ax2 = ax.twinx()
ax2.plot(serie_mdc,'green', label= "barcelona")
plt.title('madrid vs barcelona')
plt.xlabel('data')
ax.legend(loc=2)
ax2.legend(loc=0)
plt.grid(True)
plt.show()
```

> Parece que la movilidad se recupero con mayor velocidad y potencia en catalunña que en madrid. Podria ser que el hecho de ser la mitad de habitantes haga mas facil la recuperacion.
