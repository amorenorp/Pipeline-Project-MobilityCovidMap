### Presentacion de datos

Trabajaremos con los informes de tendencias de movilidad facilitada por Apple. Los informes se publican diariamente y reflejan las solicitudes de indicaciones en Mapas de Apple.
En el siguiente enlace podremos descargar los datos al completo https://covid19.apple.com/mobility con los cambios diarios en las solicitudes de indicaciones por tipo de transporte para todos los países, regiones, subregiones y ciudades disponibles.
Los datos del 11 y 12 de mayo no están disponibles, por lo que aparecerán como columnas en blanco en el conjunto de datos.

Estos datos se generan contando la cantidad de solicitudes realizadas a Mapas de Apple para obtener indicaciones en países, regiones, subregiones y ciudades concretas. Los datos que se envían desde los dispositivos de los usuarios al servicio de Mapas están asociados con identificadores rotativos aleatorios para que Apple no tenga un perfil de los movimientos y búsquedas a nivel individual. La disponibilidad de datos de un país, región, subregión o ciudad en particular se basa en una serie de factores, incluidos los umbrales mínimos para las solicitudes de indicaciones al día.

Acerca de estos datos
El archivo CSV y los gráficos de este sitio muestran el volumen relativo de las solicitudes de indicaciones por país, región, subregión o ciudad respecto al volumen de referencia del 13 de enero de 2020. Los días comienzan y terminan a medianoche, hora del Pacífico. Entendemos el término “ciudades” como grandes áreas metropolitanas cuyos límites geográficos son siempre los mismos para el conjunto de datos. En muchos países, regiones, subregiones y ciudades, el volumen relativo ha aumentado desde el 13 de enero, de acuerdo con un uso normal y estacional de Mapas de Apple. Cuando utilices estos datos, es importante normalizar cómo influyen los días de la semana. Los datos que se envían desde los dispositivos de los usuarios al servicio de Mapas están asociados con identificadores rotativos aleatorios para que Apple no tenga un perfil de los movimientos y búsquedas a nivel individual. Mapas de Apple no tiene información demográfica sobre nuestros usuarios, por lo que no podemos hacer ninguna afirmación sobre la representatividad de su uso respecto a la población en general.

###  Valores nulos

Al tratarse de tendencias la presencia de un 0 puede afectar de forma importante al analisis de los datos 

```python
nulos_df=df.isnull().sum()
nulos_df
```

Observamos como los valores nulos coinciden en cantidad en las series temporales.Por lo que procedemos a visualizar los correspondientes.

```python
listado_nulos = pd.isna(df['2021-01-26'])
listado_nulos_df = df[listado_nulos]
listado_nulos_df
```

Coprobacion de relevancia de los registros sospechosos mediante comparacion de duplicidades.

1 - generacion copia del df inicial contando con la unica columna de region como punto de referencia.

```python
copia_df = df.drop(['geo_type', 'transportation_type', 'alternative_name',
       'sub-region', 'country'],axis=1,inplace=False).copy()
```

2 - aplicacion del metodo .isna sobre una unica columna columna sospechosa.

```python
lista_nulos = pd.isna(copia_df['2021-01-26'])
listado_nulos_df = copia_df[listado_nulos]
listado_nulos_df
```

3 - comparacion mediante duplicidades de registro. 

```python
listado_nulos_df.isna().duplicated().sum()
```
